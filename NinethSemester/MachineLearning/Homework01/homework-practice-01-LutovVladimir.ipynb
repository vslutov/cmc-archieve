С помощью пакета matplotlib или plotly постройте графики времени работы в зависимости от размера данных. __Графики должны выглядеть опрятно!__ То есть должны быть подписаны оси, названия графиков, и т.д. Например, ниже представлены хороший и плохой графики:

```python
import functions
import functions_vectorized
import generators
import numpy as np

from scipy.signal import medfilt
from time import time
from matplotlib import pyplot as plt

def performance_test(simple, vectorized, gen, title, xlabel, max_size=500, count=20, first_label=u"Невекторизованная", second_label=u"Векторизованная"):
    time_non_vectorized, time_vectorized = [], []
    
    for args in gen(max_size):
        vect_args = tuple(np.array(arg) for arg in args)
        
        start = time()
        for c in range(count):
            simple(*args)
        time_non_vectorized.append((time() - start) / count)
        
        start = time()
        for c in range(count):
            vectorized(*vect_args)
        time_vectorized.append((time() - start) / count)
    
    input_size = list(range(max_size - len(time_vectorized), max_size))
    time_non_vectorized = medfilt(time_non_vectorized, 5)
    time_vectorized = medfilt(time_vectorized, 5)
    
    plt.plot(input_size, time_non_vectorized, label=first_label)
    plt.plot(input_size, time_vectorized, label=second_label)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(u"Время (с)")
    plt.grid()
    plt.legend()
    plt.show()
```

```python
from scipy.spatial import distance

performance_test(functions.prod_non_zero_diag, functions_vectorized.prod_non_zero_diag, generators.prod_non_zero_diag, u"Произведение ненулевых элементов на диагонали", u"Ширина и высота матрицы")
performance_test(functions.are_multisets_equal, functions_vectorized.are_multisets_equal, generators.are_multisets_equal, u"Равенство мультимножеств", u"Размер мультимножеств")
performance_test(functions.max_after_zero, functions_vectorized.max_after_zero, generators.max_after_zero, u"Максимум среди элементов после нулей", u"Размер массива")
performance_test(functions.convert_image, functions_vectorized.convert_image, generators.convert_image, u"Преобразование изображения", u"Ширина и высота изображения (px)", max_size=30)
performance_test(functions.run_length_encoding, functions_vectorized.run_length_encoding, generators.run_length_encoding, u"RLE кодирование", u"Размер массива", max_size=1500)
performance_test(functions.pairwise_distance, functions_vectorized.pairwise_distance, generators.pairwise_distance, u"Расчет расстояний по всем парам объектов", u"Количество объектов", max_size=30)
performance_test(distance.cdist, functions_vectorized.pairwise_distance, generators.pairwise_distance, u"Расчет расстояний по всем парам объектов", u"Количество объектов", max_size=100, first_label=u"scipy.spatial.distance.cdist")
```

### Задача 1: пропуски в данных

Загрузите таблицу data.csv. Проверьте, есть ли в ней пропуски (значения NaN). Замените все пропущенные значения на среднее значение этого признака по столбцу.

```python
import pandas as pd
import numpy as np
data = pd.read_csv("data.csv")
data = data.fillna(data.median())
scores = pd.read_csv("scores.csv", names=["score"])
```

### Задача 2: кросс-валидация для исходных данных
__(0.5 балла)__

Загрузите файл scores.csv и протестируйте, как линейная регрессия предсказывает ответ сейчас (с помощью кросс-валидации).

*Hint: воспользуйтесь sklearn.linear_model и sklearn.model_selection.*

```python
from sklearn.model_selection import train_test_split, KFold
from sklearn import linear_model, metrics
import csv
from matplotlib import pyplot as plt

kf = KFold(n_splits=3, random_state=123, shuffle=True)
reg = linear_model.LinearRegression()

def rmse(y_true, y_pred):
    return np.sqrt(metrics.mean_squared_error(y_true, y_pred))
    
def show_quality(y_true, y_pred):
    print("RMSE:", rmse(y_true, y_pred))
```

```python
X = data.as_matrix()
y = scores["score"].as_matrix()

for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    reg.fit(X_train, y_train)
    y_pred = reg.predict(X_test)
    show_quality(y_test, y_pred)
```
